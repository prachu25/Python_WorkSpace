```python
import pymysql
```

```python
connection = pymysql.connect(
    host=('HOST_NAME'),
    port=('PORT_NUMBER'),
    user=('USER_NAME'),
    password=('PASSWORD'),
    database=('DATABASE_NAME')
    )
print('connected successfully')
```

```python
cursor = connection.cursor()
```

```python
try:
  ano = int(input('Enter account number: '))
  cnt = cursor.execute(f"delete from accounts where accno={ano} ")
  connection.commit()

  if cnt > 0:
    print('account closed successfully..')
  else:
    print("account not found")

except ValueError:
  print('please enter a valid details..')
```
