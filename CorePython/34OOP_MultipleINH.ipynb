**Multiple Inheritance**

**When a child class inherits from more than one parent class, it is called Multiple Inheritance.**

One child → Multiple parents

```python
# syntax
class Parent1:
  pass

class Parent2:
  pass

class Child(Parent1,Parent2):
  pass

```

```python
# pass in Python is a keyword that means “do nothing.”
# It is used as a placeholder when a statement is required but you don’t want to write code yet.

def my_function():
  pass
```

```python
class One:
  def calsquare(self,n):
    print(f'square is {n**2}')
```

```python
class Two:
  def calSum(self, n1,n2):
    res = n1 + n2
    print(f'sum is {res}')
```

```python
class Three(One,Two):       # inhert class one and class two
  def calCube(self,n):
    cube = n**3
    print(f'cube is {cube}')
```

```python
obj = Three()
```

```python
obj.calCube(2)    # own method
obj.calsquare(3)   # inherit from class One
obj.calSum(1,2)  # inherit class from Two
```

```python
# Example Combine Skills
class Singer:
  def sing(self):
    print('his very good singer ')

class Dancer:
  def dance(self):
    print('his a very good Dancer')

class Performance(Dancer,Singer):
  pass

class Devloper(Performance):
  pass

obj = Devloper()
obj.sing()
obj.dance()
```

```python
obj = Performance()
obj.dance()
obj.sing()
```
