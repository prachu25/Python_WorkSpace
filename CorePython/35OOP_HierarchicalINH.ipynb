**Hierarchical Inheritance**

One Parent class                                
Many child class                      
All child classes inherit from the same parent class


```python
# SYNTAX
class Parent:
    pass

class Child1(Parent):
    pass

class Child2(Parent):
    pass
```

```python
# Parent class
class Vehicles:
  def start(self):
    print('vehical start..')

  def stop(self):
    print('vehicle stop')

  def derive(self,speed=45):
    print(f'vehicle drive at {speed} km/h')

class Car(Vehicles):
  def bmw(self):
    print('BMW car..')

class Jeep(Vehicles):
  def luxury(self):
    print('luxury jeep..')
```

```python
obj = Vehicles() # object creation parent class
obj.derive()
obj.start()
obj.stop()
```

```python
c = Car()
c.bmw()   # own
c.start()
c.derive()
c.stop
```

```python
# Parent class
class Animal:
    def sound(self):
        print("Animals make sound")

# Child class 1
class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Child class 2
class Cat(Animal):
    def meow(self):
        print("Cat meows")

# Creating objects
d = Dog()
c = Cat()

d.sound()   # Inherited method
d.bark()

c.sound()   # Inherited method
c.meow()
```
