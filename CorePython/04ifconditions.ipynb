**find out if the number is odd or even**

```python
num = int(input('Enter a number: '))
if num%2==0:
  print('number is even')
else:
  print('number is odd')
```

```python
n = int(input('Enter a number: '))
if n==0:
  print('number is zero/natural')
else:
  if n>0:
    print('number is positive')
  else:
    print('number is negative')
```
