####**Dictionary** - dictionary (dict) is a data structure that store in key:value pairs

```python
dic = {}
print(type(dic))
```

```python
student = {
    "name":"joe",
    "age":22,
    "course":"python",
    "city":"mumbai"
}
print(student)
```

```python
student["course"] = ["java","React js","spring boot"]
print(student)
```

```python
# zip() combines two lists into pairs (key, value).
# dict() converts those pairs into a dictionary.

k = ['patnm','age','city']   # keys
v = ['ishan',23,'mumbai']    # values

combine = dict(zip(k,v))
print(combine)
```

```python
combine['age'] = 45          # age update
print(combine)
```
