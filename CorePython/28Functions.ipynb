**Defalut value** - A default value is a value that a function uses automatically if no value is given when calling it.

```python
def greet(name='Guest'):
  print('Hello',name)

greet('jon')   # give value
greet()     # uses default value
```

```python
# *names is used to take multiple values as input.
#  It collects all values into a tuple

def congrats(*names):
  print(names)
  for nm in names:
    print(f"congrats {nm}")

congrats('will','mike','max')
```

What *args means

*args allows you to pass any number of values

All values get stored in a tuple

```python
def total(*args):
  print(f'sum of values is: {sum(args)}')
  print(f'largest values is: {max(args)}')
```

```python
total(2,3,4)
total(2,3,4,6,3,3)
```

What **kwargs means

**kwargs lets a function take any number of keyword arguments

**It stores them in a dictionary**

```python
def display(**kwargs):
  for key, value in kwargs.items():
    print(f"key: {key}, value: {value}")
```

```python
display(name="rocky", age=21, city="Nagpur")
```

```python
def display(**kwargs):
  for key, value in kwargs.items():
    print(f"{key} : {value}")
```

```python
display(name="rocky", age=21, city="Nagpur")
```

```python
# Multiply all Numbers
def multiply(*nums):
  result = 1
  for n in nums:
    result *=n
  print('Multiplication: ', result)
```

```python
multiply(2,3,4,5)   # 2*3*4 = 24    24*5 => 120
```

```python
# Find min
def small(*nums):
  print('smallest value: ', min(nums))

small(10,2,33,4,5)
small(10,2,-7)
```

```python
# Count values
def count_values(*data):
  print("Total values: ",len(data))

count_values(1,2,22,9)        # 4 values here
```
