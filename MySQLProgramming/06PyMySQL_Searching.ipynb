```python
import pymysql
```

```python
connection = pymysql.connect(
    host=('HOST_NAME'),
    port=('PORT_NUMBER'),
    user=('USER_NAME'),
    password=('PASSWORD'),
    database=('DATABASE_NAME')
    )
print('connected successfully')
```

```python
cur = connection.cursor()
```

```python
try:
  no = int(input('Enter account number to search: '))
  cur.execute(f'select * from accounts where accno = {no}')
  data = cur.fetchone()

  if data:
    print(data)
  else:
    print('account does not exist')

except ValueError:
  print('invalid input.. ')
```

```python
typ = input('Enter account Type: ')
cur.execute(f"select * from accounts where acctype='{typ}'")
data = cur.fetchall()

for i in data:
  print(i)
```

```python
comp=input('Enter company : ')
cur.execute(f"select * from cars where company='{comp}'")
data=cur.fetchall()

for i in data:
  print(i)
```

```python
dis=input('Enter keyword relating to the problem : ')
cur.execute(f"select * from patients where disease like '%{dis}%'")
data=cur.fetchall()
print(data)
```
