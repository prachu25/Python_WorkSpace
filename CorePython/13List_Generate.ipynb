```python
lst=[3,12,45,9,15,26,13]
sq=[]

for n in lst:
  sq.append(n*n)

print(sq)
```

```python
lst = [2,4,5,6,7,8]
cubes = [n**2 for n in lst]
print(cubes)
```

```python
num = [1,2,3,4,5,6,7,8,9,12,1,3,14,15,16,17,18,19,20]

even = [n for n in num if n%2==0]
print(even)
```

```python
odd = [n for n in num if n%2==1]
print(odd)
```

```python
nms = ['will','lucus','max','11','stevth','mike']
u = [nm.upper() for nm in nms]
print(u)
```

```python
cap = [nm.capitalize() for nm in nms]
print(cap)
```

```python
no = [1,5,6,3,0,-9,-6,-4]
neg = [x for x in no if x<0]
print(neg)
```
