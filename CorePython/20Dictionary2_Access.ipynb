```python
cars = {
    "car1": {
        "brand": "Swift",
        "year": 2022,
        "company": "Maruti Suzuki"
    },
    "car2": {
        "brand": "i20",
        "year": 2023,
        "company": "Hyundai"
    },
    "car3": {
        "brand": "Fortuner",
        "year": 2021,
        "company": "Toyota"
    }
}

print(cars)
```

```python
for key in cars:
  print(cars.get(key))
```

```python
for key,value in cars.items():
  print(key,value)
```

```python
cars = [
    {"brand": "Swift", "year": 2022, "company": "Maruti"},
    {"brand": "i20", "year": 2023, "company": "Hyundai"},
    {"brand": "Fortuner", "year": 2021, "company": "Toyota"}
]

for car in cars:
  print(car['brand'],car['year'])
```

```python
cars = {
    "car1": "Swift",
    "car2": "i20",
    "car3": "Fortuner"
}

for car in cars.keys():
    print(car)        # key
    print(cars[car])  # value
```

```python
cars = {
    "car1": "Swift",
    "car2": "i20",
    "car3": "Fortuner"
}

for value in cars.values():
  print(value)
```

```python
for key in cars.keys():
  print(key)
```

```python
for key, value in cars.items():
    print(key, value)
```
