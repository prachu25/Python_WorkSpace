```python
import pymysql
```

```python
connection = pymysql.connect(
    host=('HOST_NAME'),
    port=('PORT_NUMBER'),
    user=('USER_NAME'),
    password=('PASSWORD'),
    database=('DATABASE_NAME')
    )
print('connected successfully')
```

```python
cursor = connection.cursor()
```

**INSERT IN ACCOUNTS**

```python
no = int(input("Enter account number: "))
nm = input('Enter account name: ')
typ = input('Enter account type:')
balance = float(input('Enter balance: '))
```

```python
cursor.execute(f"insert into accounts values({no},'{nm}','{typ}',{balance})")
```

```python
connection.commit()
```

```python
connection.close()
```
