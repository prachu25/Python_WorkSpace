```python
# Student Mark System
marks = int(input('Enter your Marks: '))

try:
  if marks < 0 or marks > 100:
    raise ValueError('Invalid marks enter')
  elif marks >=35 :
    print('You Pass')
  else:
    print('Fail')

  print('marks accepted')

except ValueError as e:
  print('Error: ', e)
```

```python
# Online Shopping (Stock Check)

stock = 10
order = int(input('Enter Quantity: '))

try:
  if order > stock:
    raise Exception   # direct call, no message here
  print('order placed Successfully..')
except Exception:
  print('Error: Product out of Stock')
```

```python
# Zomato Order Example

restaurant_open = True
delivery_boy_available = False
order_amount = 250

try:
  if not restaurant_open:
    raise Exception

  if not delivery_boy_available:
    raise Exception

  if order_amount < 100:
    raise Exception

  print("Order placed successfully! Your food is on the way..")

except Exception:
  print('sorry! order cannot be delivered right now. ')
```

```python
# Movie Sets Booking

available_seats = 5

try:

  tickets = int(input('Enter Number of Ticket: '))
  payment = input('Payment Successful / (Yes/No): ').lower()

  if tickets > available_seats:
    raise ValueError

  if payment!='yes' :
    raise RuntimeError

  print('Ticket book Successfully! Enjoy the movie')

except RuntimeError:
  print('Invalid input..')

except ValueError:
  print("Sorry! Not enough seats available.")

except Exception:
  print('Payment Failed. try again..')
```
