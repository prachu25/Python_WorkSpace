It is used to access the object’s methods and variables inside the class.

```python
class Person:
  def welcome(self):
    print('welcome to python')

  def thanks(self,nm):
    print(f"thanks {nm} for joining pythoncource")
```

```python
class Calculator:
  def calcsquare(self,n):
    sq = n**2
    print(f'square of {n} is {sq}')

  def multiply(self,x,y):
    res = x*y
    print(f'multiplication of {x} and {y} is {res}')

  def add(self,x,y):
    res = x + y
    print(f'addition of {x} and {y} is {res}')
```

```python
c = Calculator()   # object creation in python
```

```python
c.multiply(5,7)
c.add(2,3)
c.calcsquare(7)
```

```python
# car
class Car:
  def start(self):
    print('car is started')

  def drive(self,speed = 40):
    print(f'car is running at {speed} km/h')

  def stop(self):
    print('car is sopped')
```

```python
bmw = Car()            # object creation
```

```python
bmw.start()
bmw.drive()      # default value use
bmw.drive(60)   # giving a value
bmw.stop()
```

```python
# Zomato

class Zomato:
  def order_food(self, *args):
    print(f'order placed for {args}')

  def track_order(self,status):
    print(f'your order {status}')

  def cancle_order(self,*args):
    print(f'{args} order canclled')

  def make_payment(self,amount):
    print(f"Payment of ₹{amount} successful")
```

```python
z = Zomato()
z.order_food('pizza','burger','fries','cake','ice-cream','choco-chips')
z.track_order('out of dilivery')
z.make_payment(6900)
```
