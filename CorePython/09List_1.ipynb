## Python List â€“

1. Ordered collection (maintains insertion order)

2. Allows duplicate values

3. Mutable (can be modified after creation)

4. Supports indexing (position-based access)

5. Supports negative indexing

6. Allows multiple data types in one list

7. Dynamic size (can grow or shrink)

8. Supports nested lists (list inside a list)

9. Not sorted by default

10. Supports slicing (can access a range of elements)




### ***Add elements***
1. list.append(x)        
2. list.extend(iter)      
3. list.insert(i, x)      

### ***Remove elements***
1. list.remove(x)            
2. list.pop(i)                 
3. list.pop()                 
4. list.clear()             

### ***search/Count***
1. list.index(x)       
2. list.count(x)    

### ***Sorting/Reverse***
1. list.sort()         # ascending sort
2. list.sort(reverse=True)  # descending
3. list.reverse()
