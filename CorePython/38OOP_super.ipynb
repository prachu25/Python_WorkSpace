**SUPER**                             
**super() is used to call the parent class method or constructor from the child class.**

Why We Use super()?

✅ To reuse parent code                
✅ To avoid rewriting same code          
✅ To call parent constructor (__ init __)



```python
# EXAMPLE 1
class Parent:
  def show(self):
    print('This is Parent')

class Child(Parent):
  def display(self):
    super().show()       # using super keyword print parent class method
    print('This is a child class')

c = Child()
c.display()
```

```python
# EXAMPLE 2
class Mike:
  def __init__(self,name):
    print(f'your name is {name} and length is  {len(name)}')
```

```python
class Will(Mike):
  def __init__(self):
    super().__init__('max')
    print('derived class Will constructor running')

obj = Will()
```
