**Arithmetic Operators**

```python
x = 9
y = 13

# arithmetic operators demo
print(x + y)
print(x - y)
print(x * y)
print(x / y)

# floor division
print(y//x)

# Modulus
print(y%x)
print(27%4)

# Power
print(2**3)
print(7**2)
```

**Assignment Operators**

```python
x=9
x+=3   # x=x+3
x-=2
x*=5
x/=2
x//=5
x%=2

print(x)
```

**Logical Operators**

```python
print(10>5)
print(7<5)

print()
print(10>5 and 1<5)     # True + True = True
print(10>5 and 7<5)     # True + Flase = Flase

print(10>5 or 10<4)     # True + Flase = True
print(10>7 or 10>5)     # True + True  = True
print(10>15 or 15<5)    # Flase + Flase = Flase
```

**Membership operators**

```python
print('j' in 'jos')
print('k' not in 'kite')

lst =['jos','max','mike','will','eleven','nancy']
print('max' in lst)
```


**Ternary Operator**   (value_if_true if condition else value_if_false)

```python
# Using If-else
marks = 35
if(marks >=35):
  print('pass')
else:
  print('fail')

print()
# Using Ternary
marks = 25
result = 'pass' if marks>=35 else 'fail'
print(result)
```

```python
# Ternary Example1
is_login = False
message = 'Welcome Back!' if is_login else 'Please login'
print(message)
```

```python
# Ternary Example2
num = int(input('Enter a number '))
abs = num if num>0 else -num
print(abs)
```

```python
# Ternary Example3
no = int(input('Enter a Number '))
num_even_or_odd = 'Even' if no%2==0 else 'Odd'
print(f'number is {num_even_or_odd}')
```
