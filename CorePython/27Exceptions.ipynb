
# **Exception Handling**

```python
try:
  n = int(input('Enter a number: '))
except:
  print('Invalid input.. pls enter a number')
  n=0

sq = n*n
print(f"square of {n} is {sq}")
```

```python
try:
  first = int(input('Enter a first number: '))
  second = int(input('Enter  a Second number: '))
  res = first/second

  print(f"the soln is {res}")


except ValueError:
  res=0
  print('invalid input.. please enter a number')
except ZeroDivisionError:
  res=0
  print('cant divided by zero')
```
