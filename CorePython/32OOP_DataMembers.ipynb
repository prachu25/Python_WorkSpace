**data member is a variable of the class whose values can be accessed across all functions of the class**

```python
class School:
  scnm = 'kidg school'   # Class variable (shared by all objects, accessed using self)

  def show(self,scnm):     # Local variable (inside method)
    print(f'local - school name is {scnm}')
    print(f'global- school name is {self.scnm}')

  def display(self):
    print(self.scnm)
```

```python
obj=School()
obj.show("zp school")
```

```python
# Simple Message ENCRYPTION
# (Convert each letter to next letter).
# Input: abc   => Output: bcd

msg = input('Enter message: ')
new_msg = ""

for ch in msg:
  new_msg += chr(ord(ch) + 1)   # ord Ascii

print('Encryption: ',new_msg)
```

```python
# DECRYPTION
msg = input("Enter encrypted message: ")
decrypted = ""

for ch in msg:
    decrypted += chr(ord(ch) - 1)

print("Decrypted message:", decrypted)
```
