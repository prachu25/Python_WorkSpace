**A lambda function** in Python is a small, one-line anonymous function (a function without a name) used to do simple tasks quickly.

```python
def calsquare(x):               # x**2 â†’ square power(2)
  sq = x **2                    # x**3 -> cube  power(3)
  return sq
```

```python
calsquare(2)
```

```python
# square
def sq(n):
  return n*n

print(sq(6))
res = sq(8)
print(res)
```

**lambda parameters:return value**



```python
sq = lambda x:x**2

print('square is ', sq(7))
```

```python
add = lambda a,b:a+b

print(f'addition is {add(2,3)}')
```

QUE. Write a lambda to check if a number is even or odd.

```python
# WITH LAMBDA
check_even_odd = lambda n: 'even' if n%2==0 else 'odd'

print(check_even_odd(3))
```

```python
def odd_even_check(x):         # WITHOUT LAMBDA
  if x%2==0:
    return ' even'
  else:
    return 'odd'

print(odd_even_check(8))
```

```python
max_num = lambda a,b: a if a > b else b

print(max_num(3, 8))
```

```python
# double of the given number
double_num = lambda x: x*2
print(double_num(5))
```

```python
# sub the given number by 2
sub_by_2_num = lambda x: x-2
print(sub_by_2_num(6))
```

```python
# multiplication of x and y
mul_num = lambda x,y: x*y
print(mul_num(5,7))
```

```python
# check positive or negative number
check_pos_neg = lambda x: 'positive' if x > 0 else ('neutral' if x ==0  else 'negative' )

print(check_pos_neg(0))
print(check_pos_neg(-7))
print(check_pos_neg(55))
```

```python
# half of the given number
half = lambda x:x/2
print(half(4))
```

```python
def adding(*args):        # using args
  return sum(args)

print(adding(2,3,1))
```

```python
# convert it into  uppercase
upper_case_convert = lambda name:name.upper()

print(upper_case_convert('prachi'))
print(upper_case_convert('mike'))
print(upper_case_convert('max'))
```

```python
# length of the string
length_of_string = lambda nm:len(nm)

print(length_of_string('parchi'))
print(length_of_string('max'))
print(length_of_string('will'))
```

```python
join_string = lambda x,y:x + " " + y

print(join_string('focus','academy'))
```
