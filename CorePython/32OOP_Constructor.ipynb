**Constructor** is a special function in the class **that runs automatically** when the **object is created** to perform some initialization activities.

it is **used to initialize the object's data** (variables).

The constructor in Python is written using *__init __*().

```python
# syntax
class ClassName:
    def __init__(self):
        # initialization code
        print('i am constructor')
```

```python
# when object is created constructor runs automatically
obj = ClassName()      # object creation
```

```python
# constructor with variables
class Student:
  def __init__(self,name,age):
    self.name = name
    self.age = age

s1 = Student('john',22)

print(s1.name)
print(s1.age)
```

```python
# Default value
class Car:
  def __init__(self,brand = 'TATA'):
    self.brand = brand


c1 = Car()              # use default value
print(c1.brand)

c2 = Car('bmw')          # use given value
print(c2.brand)
```

```python
class Student:
  def __init__(self,name = 'john'):
    print(f'welcome {name} to the python world')
```

```python
s = Student('prachi')
s = Student()
```

```python
# Constructor with Multiple Parameters
class Student:
  def __init__(self,nm,ag,stream):
    self.name = nm
    self.age = ag
    self.branch = stream


s = Student('john',19,'IT')
print(s.name, s.age, s.branch)
```

```python
# Default + Parameterized Together
class Mobile:
  def __init__(self,brand = 'Apple', price = 1999):
    self.brand = brand
    self.price = price


m1 = Mobile()
m2 = Mobile('samsung',8000)

print(m1.brand,m1.price)
print(m2.brand,m2.price)
```

```python
# Using user input in constructor
class Person:
  def __init__(self,name):
    self.name = name

name = input('Enter your name: ')
p = Person(name)
print('hello',p.name)
```

```python
# constructor with Calculation
class Rectangle:
  def __init__(self,l,b):
    self.area = l*b

r = Rectangle(5,3)
print('Area',r.area)
```

```python
# constructor with method
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def show(self):
        print(self.name, self.salary)

e1 = Employee("Amit", 50000)
e1.show()
```

```python
# constructor with list
class Shopping:
  def __init__(self,item):
    self.item = item

s = Shopping(['shoes','shirts','watch','books'])
print(s.item)
```

```python
# constructor with *args (we can pass multiple args here bcz *args)
class Numbers:
  def __init__(self,*nums):
    self.total = sum(nums)

n = Numbers(1,2,3)
print(n.total)
```

```python
# Constructor Using *args
class Details:
  def __init__(self,*args):
    self.info = args

  def show(self):
    print(self.info)

d = Details('prachi',22,'nagpur','python','java')
d.show()
```

```python
# Constructor Using **kwargs   (**kwargs strore -> in key : value pair)
class Details:
  def __init__(self,**kwargs):
    self.info = kwargs

  def show(self):
    for k,v in self.info.items():
      print(k , ':' , v)

d = Details(name = 'prachi',age = 22,city= 'nagpur',cource = 'java')
d.show()
```
