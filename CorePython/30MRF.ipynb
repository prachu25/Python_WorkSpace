Map - It will take all elements from the collection

```python
numbers = [1,2,3,4,5]
sq = list(map(lambda x:x**2 ,numbers))
print(sq)
```

```python
names=['england','germany','france','italy']
ucase = list(map(lambda nm:nm.upper(),names))
print(ucase)
```

```python
len_of = list(map(lambda x:len(x),names))
print(len_of)
```

**filter()** is a Python function that returns only those elements from a list (or iterable) that meet a given condition.

```python
nums = [1,2,3,4,5]

even_nums = list(filter(lambda x:x%2==0,nums))
print(even_nums)
```

```python
names=['microsoft','volkswagen','bmw','skoda','oracle','openai','copilot','sohamglobal']
res=list(filter(lambda x:len(x)>8,names))
print(res)
```

**reduce(): Combines all elements of an iterable into one value using a function.**                                      

 it return 1 summary result

```python
from functools import reduce

lst = [1,2,3,4,5]
sum = reduce(lambda x,y: x+y ,lst)
print(sum)
```

**Short explanation:**

**from functools import reduce** → imports the reduce() function.

**lst = [1,2,3,4,5]** → creates a list of numbers.

**reduce(lambda x, y: x + y, lst)** → adds numbers step by step.
(1+2=3, 3+3=6, 6+4=10, 10+5=15)

Result **(15) is stored in sum**.

print(sum) → prints 15.
