Remove duplicate values from a list

```python
numbers = [1,2,2,2,3,4,4,5,5,6,7,7,7,8,8]
unique_numbers = set(numbers)   # Sets automatically remove duplicates
print(unique_numbers)
```

```python
visited=set()
itemcodes=[1,2,3,2,1,4,3,5,2]

for item in itemcodes:
  if item not in visited:
    print(f'processing orders for {item}')
    visited.add(item)
    print(visited)
```

## ***Frozenset - immutable set***

**Normal set:**
You can add or remove elements.

**frozenset:**
Once it is created, you cannot change it (it is immutable).


```python
fs=frozenset([1,2,3,4,5,2])
print(fs)
```

Difference between set and frozenset

```python
s = {1,2,3}
s.add(4)           # allowed IN  SET
print(s)
```

```python
fs = frozenset([1, 2, 3])
fs.add(4)
print(fs)    # Error (cannot modify)   frozen set ---> immutable set
```
