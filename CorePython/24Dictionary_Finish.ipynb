```python
company={
    "IT":{
        "manager":"joe",
        "employees":["rishi","sneha","nupur","javed"]
    },
    "HR":{
        "manager":"sam",
        "employees":["megha","manali","rita"]
    }
}
```

```python
print(company['IT'])
```

```python
print(company['IT']['manager'])
```

```python
print(company['IT']['employees'])
```

```python
print(company["IT"]["employees"][0])   # on 0 index => rishi
```

```python
for nm in company['IT']['employees']:
  print(nm.upper())
```

```python
for nm in company['HR']['employees']:
  print(nm.capitalize())
```

```python
for dept,details in company.items():
  count_emp = len(details['employees'])
  print(f'{dept} has {count_emp} employees')
```

```python
count = len(company["HR"]["employees"])
print("HR employees:", count)
```
