```python
import pymysql
```

```python
connection = pymysql.connect(
    host=('HOST_NAME'),
    port=('PORT_NUMBER'),
    user=('USER_NAME'),
    password=('PASSWORD'),
    database=('DATABASE_NAME')
    )
print('connected successfully')
```

```python
cur = connection.cursor()
```

**DEPOSITE MONEY IN ACCOUNTS**

```python
no = int(input('Enter a account number: '))
amt = float(input('Enter a deposit amount: '))

cnt = cur.execute(f"update accounts set balance = balance+{amt} where accno = {no} ")
connection.commit()

if(cnt > 0):
  print('amount deposited successfully..')
else:
  print('account does not exist')
```

**WITHDRAW MONEY IN ACCOUNTS**

```python
no = int(input('Enter a account number: '))
amt = float(input('Enter a withdraw amount: '))

cnt = cur.execute(f"update accounts set balance = balance-{amt} where accno = {no} ")
connection.commit()

if(cnt > 0):
  print('amount withdraw successfully..')
else:
  print('account does not exist') # 70000 doesnt exit bezx we apply contraints at least after withdrw 500 rs left must
                                  # in that case user try to withdraw all amount that why it showing error
                                  # for that apply try except to show insufficient amount
```
