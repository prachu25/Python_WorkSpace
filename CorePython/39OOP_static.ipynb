**- A method that does NOT use self**   
**- It does NOT depend on object data**  
**- we can Directly call from the class**

```python
class Student:

  @staticmethod               # static method doesn't need the object
  def ispass(marks):
    return marks >= 35

  def thanks(self):           # normal method , needs a object
    print('thanks for your visit')
```

```python
print(Student.ispass(35))
print(Student.ispass(20))
```

```python
obj = Student()
# print(obj.ispass(45))
obj.thanks()
```
