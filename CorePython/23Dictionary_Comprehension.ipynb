*What is enumerate()?*
---
enumerate() adds a counter (index) to an iterable and returns both index and value while looping.



```python
emps = ['Jos Buttler','Pat Cummins','Joe Root','Steve Smith','Ben Stokes']

for index, value in enumerate(emps):
    print(f"{index} and {value}")
```

emps is a list of names.

enumerate(emps) gives 2 things in each loop:

index → position number

value → actual item from the list

```python
for i in range(len(emps)):          # Without enumerate
    print(i, emps[i])
```

```python
for i, name in enumerate(emps):
    print(i, name)
```

```python
for i, name in enumerate(emps,start=1):  # index start from = 1
    print(i, name)
```

enumerate() is used when you need both:

index

value

while looping through a list.
