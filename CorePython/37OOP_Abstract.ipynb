**Abstract**

Hiding implementation details
Showing only essential features.


1. **can not be used to create a object**
2. contain at least one abstract method
3. child class must be implement that method.
4. **@abstractmethod** is a decorator used inside an abstract class.
5. **This method as no body.**
6. **child class must be implement it.**


1. **from abc import ABC, abstractmethod**

ABC → Used to create an Abstract Base Class.                           
abstractmethod → Decorator to make a method compulsory in child class

2. **class Animal(ABC):**                
Animal is an Abstract class can not create a object directly.

3. **@abstractmethod**                   
**def sound(self):**                                  
pass

4. @abstractmethod makes sound() compulsory                            
Child class must define this method
5. **class Dog(Animal):**             
Dog inherits from Animal

```python
# Abstract class

from abc import ABC, abstractmethod

class Animals(ABC):

  @abstractmethod   # decorator
  def sound(self):
    pass
   # print('kiikii..kii..')

# Child class
class Dog(Animals):

  def sound(self):
   # super().sound()      # to calling parent method
    print('bow bow.. bow .. bow')

  def sleep(self):
    print('dog is sleeping..')



child = Dog() # cild obj creation
child.sleep()
child.sound()
```
