
Slicing without copying

```python
slogan=bytearray(b'Technology is power')
print(slogan)
```

```python
slice1=slogan[:7]
print(f"regular slicing : {slice1}")
```

```python
view=memoryview(slogan)
vslice=view[2:8]
print(f"Memoryview slice : {bytes(vslice)}")
```

Memory: Space where Python stores data while a program runs.

memoryview: A tool that lets you access and modify the same memory of an object without creating a new copy.
