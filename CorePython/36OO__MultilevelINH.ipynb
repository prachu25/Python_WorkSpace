#### **Multilevel Inheritance**
**one class Inherit from another class, and that class again from another class.**
**it forms a chain of inheritance.**

Grandparent → Parent → Child


```python
# Level 1
class GrandFather:
  def show1(self):
    print('this is a GrandFather')

# Level 2
class Parent(GrandFather):
  def show2(self):
    print('this is a Parent')

# Level 3
class Child(Parent):
  def show3(self):
    print('this is child')

obj = Child()
obj.show1()
obj.show2()
obj.show3()
```

```python
# Level 1
class Person:
  def show_person(self):
    print('I am a Person')

class Student(Person):
  def show_student(self):
    print('I am a student')

class Monitor(Student):
  def show_monitor(self):
    print('I am a Monitor')

o = Monitor()
o.show_monitor()  # own
o.show_person()     # inherit
o.show_student()  # inherit
```
