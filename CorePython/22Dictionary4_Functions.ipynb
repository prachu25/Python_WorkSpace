```python
student = {
    'studentnm': "sam",
    'qualification':'mba',
    'course':'data science'
}

print(student)
```

```python
print(student.keys())
```

```python
print(student.values())
```

```python
for k in student.keys():  # print keys()
  print(k)
```

```python
for v in student.values():     # print values()
  print(v)
```

```python
print(student.items())    # print key+value pairs
```

```python
for i in student.items():
  print(i)
```

**fromkeys()** - creates a new dictionary using the items of a list as keys and assigns the **same value to all those keys**.

Example meaning:
Items in the subjects list become the keys.
0 becomes the value for every key  

ex -
subjects = ['science','maths','english']

marks = dict.fromkeys(subjects, 0)

output -
marks = {
    'science': 0,
    'maths': 0,
    'english': 0
}


```python
names = ['A','B','C']
data = dict.fromkeys(names,'Absent')

print(data)
```

```python
cars = ['ford','tiber','bmw','tata']
price = dict.fromkeys(cars,80000)

print(price)
```
