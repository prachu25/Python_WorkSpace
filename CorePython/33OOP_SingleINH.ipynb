**Single Inheritance**
**one child class inherits from one** **parent class**

Child class gets all methods of parent

Parent → Base class

Child → Derived class

```python
# SYNTAX

# class Parent:
#     # parent code

# class Child(Parent):
#     # child code
```

```python
class Parent:
  def show(self):
    print('this is a parent class method')

  def add(self,a,b):
    res = a+b
    print(res)

class Child(Parent):
  def display(self):
    print('this is a child class method')
```

```python
p = Parent()
p.add(2,1)         # parent
p.show()         # parent
```

```python
c = Child()
c.display() # child class method
c.add(2,2)  # parent class method
c.show()    # parent class method
```

```python
class Person:
  def eat(self):
    print('person can eat')

  def sleep(self):
    print('person can sleep')

  def work(self):
    print('person can work')

p = Person()
p.eat()
p.sleep()
p.work()
```

```python
class Devloper(Person):
  def code(self):
    print('devloper can code')

  def study(self):
    print('devloper can study')
```

```python
d = Devloper()
d.code()     # own
d.work()          # parent class method
d.eat()           # parent class method
d.sleep()         # parent class method
d.study()    # own
```
