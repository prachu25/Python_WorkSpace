```python
import pymysql
```

```python
connection = pymysql.connect(
    host=('DB_HOST'),
    port=('PORT'),
    user=('USER_NAME'),
    password=('PASSWORD'),
    database=('DATABASE_NAME')
    )
print('connected successfully...')
```

```python
cursor = connection.cursor()
```

```python
cursor.execute("select * from accounts")
data = cursor.fetchall()

for i in data:
  print(i)
```

```python
cursor.execute("select * from accounts")
data = cursor.fetchone()
print(data)
```

```python
data = cursor.fetchone()
print(data)
```

```python
data = cursor.fetchmany(3)
# print(data)

for i in data:
  print(i)
```

**JOINS**

```python
cursor.execute("select c.custnm, car.modelnm, car.price from customers c join cars car using(carid)")
```

```python
data = cursor.fetchall()
for i in data:
  print(i)
```

```python
cursor.execute("select * from accounts where acctype='ppf'")
data=cursor.fetchall()
for acc in data:
  print(acc)
```

```python
cursor.execute("select * from accounts where accnm like 's%' ")
data=cursor.fetchall()
for acc in data:
  print(acc)
```
